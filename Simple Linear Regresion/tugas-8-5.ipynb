```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('salary_dataset.csv')

# Memisahkan fitur (X) dan target (y)
year = data[['YearsExperience']]
salary = data['Salary']

plt.scatter(year, salary)

# Reshape salary if needed
salary = salary.values.reshape(-1, 1)

linreg = LinearRegression()
linreg.fit(year, salary)

plt.scatter(year, salary, color='red')
plt.plot(year, linreg.predict(year), color='blue')

print("Coefficients: ", )
print("Intercept: ", linreg.intercept_)

plt.show()
```
